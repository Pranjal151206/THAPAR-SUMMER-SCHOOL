```python
import time as t
import random as r
import seaborn as sns

list = []
duration = []
count = []
k = 0

for i in range (5000, 30000, 5000):
  for j in range (0, i):
    list.append(r.random())
  s = t.time()
  list.sort()
  e = t.time()
  duration.append(round(e - s, 10))
  k+=1
  list.clear()
  count.append(i)

print(duration)
print(count)

sns.lineplot(x = count, y = duration)
```

```python
names = {}

students = int(input("Enter number of students: "))
for i in range(students):
    name = input("Enter name: ")
    marks = []
    for j in range(5):
        m = float(input("Enter marks: "))  # Changed from int to float
        marks.append(m)
    names[name] = marks

max_avg = -1.0
min_avg = 101.0
top = ""
bottom = ""

for k in names:
    total = sum(names[k])  # sum of marks (float-compatible)
    ag = total / 5
    if max_avg < ag:
        max_avg = ag
        top = k
    if min_avg > ag:
        min_avg = ag
        bottom = k

print(top, "is at the top with average", max_avg)
print(bottom, "is at the bottom with average", min_avg)

```
