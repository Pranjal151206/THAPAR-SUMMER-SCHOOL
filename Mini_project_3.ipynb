```python
import random as r
Set=[-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]
subset = []
```

```python
from itertools import combinations
```

```python
counter = 0
for i in combinations(Set, 5):
  if (sum(i) == 0):
    subset.append(i)
    counter += 1
for j in subset:
  print(j)
print("total subsets = ", counter)
```

```python
counter = 0
subsets2 = []

for i in range (3, 7):
  for j in combinations(Set, i):
    subsets2.append(j)
    counter += 1

for k in subsets2:
  print(k)
print("total subsets = ",counter)
```
